```python
import pandas as pd
import os 
```

```python
file='TGI 数据源/TGI指数案例数据.xlsx'
df = pd.read_excel(file)
df.head(10)
```

```python
df.describe()
```

```python
df_usr=df.groupby('买家昵称')['实付金额'].mean().reset_index()
```

```python
unique_city_user = df.drop_duplicates(['城市','买家昵称'])
```

```python


df_usr['高客单']=df_usr['实付金额'].apply(lambda x: 1 if x<50 else 0)

merge_city_user=pd.merge(left=unique_city_user,right=df_usr,on='买家昵称',how='left')
merge_city_user.head(3)
```

```python
def high_price(tdf):
    return len(tdf[tdf['高客单_y']==1])/len(tdf)
df_city= merge_city_user.groupby('城市').apply(lambda x: high_price(x)).reset_index()
df_city.sort_values(['0'],ascending=False,inplace=True) 

df_city.head(10)
```
